# making a new column for ep count

```python
def extract_ep (txt):
    check = False
    data = ""
    for i in txt :
        if i == ")":
            check = False
            return data
        if check == True :
            data += i
        if i == "(" :
            check = True
    
```

```python
df["Episodes" ] = df["Title"].apply(extract_ep)
```

```python
df
```

```python
df["Episodes"] = df["Episodes"].str.replace("eps","")
```

# making a new column for the time stamp

```python
df.loc[0]["Title"]
```

```python
def extract_time(txt):
    check = False 
    data = ""
    for i in range(len(txt)):
        if txt[i] == ')':
            for j in range(i+1, i+20) :
                data += txt[j]

            return data 
```

```python
df["Total Time"] = df["Title"].apply(extract_time)
```

```python
df.head()
```

```python
from dateutil.relativedelta import relativedelta
from datetime import datetime
```

```python
def calculate_totalmon (period):
    try :
        start_str, end_str = period.split(" - ")
        start_date = datetime.strptime(start_str, '%b %Y')
        end_date = datetime.strptime(end_str, '%b %Y')
        r = relativedelta(end_date,start_date)
        return r.years * 12 + r.months + 1 # to include starting month
    except :
        return None 
```

```python
df['Months'] = df ['Total Time'].apply(calculate_totalmon)
```
