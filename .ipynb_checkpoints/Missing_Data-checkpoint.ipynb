```python
import numpy as np
import pandas as pd
```

# Finding Missing Data

```python
data = {
    'A' : [1,2, np.nan ,4 ,5],
    'B' : [1 ,2,3,4,5],
    'C' : [1,2,3, np.nan ,np.nan],
    'D' : [1,np.nan ,np.nan,np.nan ,5],
}

df = pd.DataFrame(data)
df
```

```python
#To find the null values 

df.isna()
```

```python
df.isna().sum()
```

```python
df.isna().any()   #cheecks if all columns have null value
```

# removing missing data 

```python
df
```

```python
df.dropna()   #row wise 
```

```python
df
```

```python
df.dropna(thresh = 3)   #rows that have atleast 3 non null value 
```

# Filling the Missing data

```python
df
```

```python
df.fillna(0)    #fills all the NAN values with 0     , column-wise
```

```python
values = {'A' : 0 , 'B':100, 'C': 300, 'D' : 400}
df.fillna(value = values)
```

```python
df
```

```python
df.fillna(df.mean())
```
