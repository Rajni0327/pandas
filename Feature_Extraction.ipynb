# making a new column for ep count

```python
def extract_ep (txt):
    check = False
    data = ""
    for i in txt :
        if i == ")":
            check = False
            return data
        if check == True :
            data += i
        if i == "(" :
            check = True
    
```

```python
df["Episodes" ] = df["Title"].apply(extract_ep)
```

```python
df["Episodes"] = df["Episodes"].str.replace("eps","")
```

```python
def extract_time(txt):
    check = False 
    data = ""
    for i in range(len(txt)):
        if txt [i] == ')':
            for j in range(i , i+18) :
                data += txt[j]

        return data 
```
