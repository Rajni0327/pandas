```python
import numpy as np
import pandas as pd
```

```python
labels = ['a','b','c']
a_list = [10 ,20,30]
arr = np.array([10,20,30])
d = {1 : 10,2 : 20,3 : 30}
```

```python
pd.Series(a_list)
```

```python
pd.Series(a_list,index = labels)
```

```python
pd.Series(arr)
```

```python
pd.Series(d)
```
